```python
import numpy as np
import pandas as pd
```

```python
df2023= pd.read_csv('2023.csv', sep='\t')
df2023
```

```python
# Remove rows where 'Institution' is NaN (assuming it indicates unnecessary rows)
df2023 = df2023.dropna(subset=['Institution'])
# Replace 'Top 0.*%' values in 'World Rank' column with sequential numbers
df2023['World Rank'] = df2023['World Rank'].replace(r'Top 0.*%', np.nan, regex=True)
df2023['World Rank'] = pd.Series(range(1, len(df2023) + 1), index=df2023.index)

# Reset index 
df2023 = df2023.reset_index(drop=True)

df2023
```

```python
df2023['Year']= 2023
df2023
```

```python
df2023.info()
```

```python
df2023.describe(include='all').T
```

```python
# Define the function to check for '>'
def has_more(value):
    return '>' in str(value)

df2023.applymap(has_more).sum()
```

```python
# Define the function to check for '+'
def has_plus(value):
    return '+' in str(value)

df2023.applymap(has_plus).sum()
```

```python
# Define the function to check for '-'
def has_dash(value):
    return '-' in str(value)

df2023.applymap(has_dash).sum()
```

```python
# change - values to Nan
for i in df2023.columns:
    df2023[i] = df2023[i].replace(['-'] , np.nan)     

df2023
```

```python
# Define the function to check for '-'
def has_dash(value):
    return '-' in str(value)

df2023.applymap(has_dash).sum()
```

```python
# convert values of columns to int & float , except columns Institution & Location
for i in df2023.columns:
    if (i == "Institution") or (i == "Location"):
        df2023[i] = df2023[i]
    else:    
        df2023[i] = pd.to_numeric(df2023[i])

df2023
```

```python
df2023.info()
```

```python
df2023.max()
```

```python
# Replace all nan values to max+1 value of their columns
for i in df2023.columns:
    if (i == "Institution") or (i == "Location"):
        df2023[i] = df2023[i]
    else:    
        df2023[i] = df2023[i].replace(np.nan, df2023[i].max() + 1)

df2023
```

```python
df2023.max()
```

```python
for i in df2023.columns:
    if i in ["Employability Rank", "Research Rank"]:
        # Apply where condition to specific columns, keeping values <= 1000 or setting to 1001
        df2023[i] = df2023[i].where(df2023[i] <= 1000, 1001)

df2023.max()
```

```python
df2023.info()
```

```python
# Assuming df2023 is your DataFrame
cols = df2023.columns.tolist()  # Convert columns to a list

# Move 'Year' column to be right after 'Location' column
cols.insert(cols.index('Location') + 1, cols.pop(cols.index('Year')))
# Reorder DataFrame with the modified columns list
df2023 = df2023[cols]
df2023
```
