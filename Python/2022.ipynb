```python
# Define the function to check for '>'
def has_more(value):
    return '>' in str(value)

df2022.applymap(has_more).sum()
```

```python
# Define the function to check for '+'
def has_plus(value):
    return '+' in str(value)

df2022.applymap(has_plus).sum()
```

```python
# Define the function to check for '-'
def has_dash(value):
    return '-' in str(value)

df2022.applymap(has_dash).sum()
```

```python
# change - values to Nan
for i in df2022.columns:
    df2022[i] = df2022[i].replace(['-'] , np.nan)     

df2022
```

```python
# Define the function to check for '-'
def has_dash(value):
    return '-' in str(value)

df2022.applymap(has_dash).sum()
```

```python
# convert values of columns to int & float , except columns Institution & Location
for i in df2022.columns:
    if (i == "Institution") or (i == "Location"):
        df2022[i] = df2022[i]
    else:    
        df2022[i] = pd.to_numeric(df2022[i])

df2022
```

```python
df2022.info()
```

```python
df2022.max()
```

```python
# Replace all nan values to max+1 value of their columns
for i in df2022.columns:
    if (i == "Institution") or (i == "Location"):
        df2022[i] = df2022[i]
    else:    
        df2022[i] = df2022[i].replace(np.nan, df2022[i].max() + 1)

df2022
```

```python
df2022.max()
```

```python
for i in df2022.columns:
    if i in ["Employability Rank", "Research Rank"]:
        # Apply where condition to specific columns, keeping values <= 1000 or setting to 1001
        df2022[i] = df2022[i].where(df2022[i] <= 1000, 1001)

df2022.max()
```

```python
# Assuming df2022 is your DataFrame
cols = df2022.columns.tolist()  # Convert columns to a list

# Move 'Year' column to be right after 'Location' column
cols.insert(cols.index('Location') + 1, cols.pop(cols.index('Year')))
# Reorder DataFrame with the modified columns list
df2022 = df2022[cols]
df2022
```
