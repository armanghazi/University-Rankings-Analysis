```python
import numpy as np
import pandas as pd
```

```python
df2021= pd.read_csv('2021.csv', sep='\t')
df2021
```

```python
df2021['Year']= 2021
df2021
```

```python
df2021.info()
```

```python
df2021.describe(include='all').T
```

```python
# Define the function to check for '>'
def has_more(value):
    return '>' in str(value)

df2021.applymap(has_more).sum()
```

```python
# Define the function to check for '+'
def has_plus(value):
    return '+' in str(value)

df2021.applymap(has_plus).sum()
```

```python
# Define the function to check for '-'
def has_dash(value):
    return '-' in str(value)

df2021.applymap(has_dash).sum()
```

```python
# change - values to Nan
for i in df2021.columns:
    df2021[i] = df2021[i].replace(['-'] , np.nan)     

df2021
```

```python
# Define the function to check for '-'
def has_dash(value):
    return '-' in str(value)

df2021.applymap(has_dash).sum()
```

```python
# convert values of columns to int & float , except columns Institution & Location
for i in df2021.columns:
    if (i == "Institution") or (i == "Country"):
        df2021[i] = df2021[i]
    else:    
        df2021[i] = pd.to_numeric(df2021[i])
df2021
```

```python
df2021.info()
```

```python
df2021.max()
```

```python
# Replace all nan values to max+1 value of their columns
for i in df2021.columns:
    if (i == "Institution") or (i == "Country"):
        df2021[i] = df2021[i]
    else:    
        df2021[i] = df2021[i].replace(np.nan, df2021[i].max() + 1)

df2021
```

```python
df2021.max()
```

```python
for i in df2021.columns:
    if i in ["Employability Rank", "Research Rank"]:
        # Apply where condition to specific columns, keeping values <= 1000 or setting to 1001
        df2021[i] = df2021[i].where(df2021[i] <= 1000, 1001)

df2021.max()
```

```python
df2021.info()
```

```python
df2021.rename(columns={'Country': 'Location'}, inplace=True)
df2021
```

```python
# Assuming df2021 is your DataFrame
cols = df2021.columns.tolist()  # Convert columns to a list

# Move 'Year' column to be right after 'Location' column
cols.insert(cols.index('Location') + 1, cols.pop(cols.index('Year')))
# Reorder DataFrame with the modified columns list
df2021 = df2021[cols]
df2021
```
