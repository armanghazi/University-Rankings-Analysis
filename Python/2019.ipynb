```python
import numpy as np
import pandas as pd
```

```python
df2019= pd.read_csv('2019.csv', sep='\t')
df2019
```

```python
df2019['Year']= 2019
df2019
```

```python
df2019.info()
```

```python
df2019.describe(include='all').T
```

```python
# Define the function to check for '>'
def has_more(value):
    return '>' in str(value)

df2019.applymap(has_more).sum()
```

```python
# Define the function to check for '+'
def has_plus(value):
    return '+' in str(value)

df2019.applymap(has_plus).sum()
```

```python
# Define the function to check for '-'
def has_dash(value):
    return '-' in str(value)

df2019.applymap(has_dash).sum()
```

```python
# change - values to Nan
for i in df2019.columns:
    df2019[i] = df2019[i].replace(['-'] , np.nan)     

df2019
```

```python
# Define the function to check for '-'
def has_dash(value):
    return '-' in str(value)

df2019.applymap(has_dash).sum()
```

```python
# convert values of columns to int & float , except columns Institution & Location
for i in df2019.columns:
    if (i == "Institution") or (i == "Location"):
        df2019[i] = df2019[i]
    else:    
        df2019[i] = pd.to_numeric(df2019[i])
df2019
```

```python
df2019.info()
```

```python
df2019.max()
```

```python
# Replace all nan values to max+1 value of their columns
for i in df2019.columns:
    if (i == "Institution") or (i == "Location"):
        df2019[i] = df2019[i]
    else:    
        df2019[i] = df2019[i].replace(np.nan, df2019[i].max() + 1)
df2019
```

```python
df2019.max()
```

```python
for i in df2019.columns:
    if i in ["Employability Rank", "Research Rank"]:
        # Apply where condition to specific columns, keeping values <= 1000 or setting to 1001
        df2019[i] = df2019[i].where(df2019[i] <= 1000, 1001)

df2019.max()
```

```python
df2019.info()
```

```python
# Assuming df2019 is your DataFrame
cols = df2019.columns.tolist()  # Convert columns to a list

# Move 'Year' column to be right after 'Location' column
cols.insert(cols.index('Location') + 1, cols.pop(cols.index('Year')))
# Reorder DataFrame with the modified columns list
df2019 = df2019[cols]
df2019
```
