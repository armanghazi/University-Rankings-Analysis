```python
import numpy as np
import pandas as pd
```

```python
df2024= pd.read_csv('2024.csv', sep='\t')
df2024
```

```python
# Remove rows where 'Institution' is NaN (assuming it indicates unnecessary rows)
df2024 = df2024.dropna(subset=['Institution'])
# Replace 'Top 0.*%' values in 'World Rank' column with sequential numbers
df2024['World Rank'] = df2024['World Rank'].replace(r'Top 0.*%', np.nan, regex=True)
df2024['World Rank'] = pd.Series(range(1, len(df2024) + 1), index=df2024.index)

# Reset index 
df2024 = df2024.reset_index(drop=True)

df2024
```

```python
df2024['Year']= 2024
df2024
```

```python
# to see the number of values equal to '-'
df2024.isin(['>']).sum()
```

```python
# Define the function to check for '+'
def has_plus(value):
    return '+' in str(value)

df2024.applymap(has_plus).sum()
```

```python
# to see the number of values equal to '-'
df2024.isin(['-']).sum()
```

```python
# change - values to Nan
for i in df2024.columns:
    df2024[i] = df2024[i].replace(['-'] , np.nan)     

df2024
```

```python
df2024.info()
```

```python
# convert values of columns to int & float , except columns Institution & Location
for i in df2024.columns:
    if (i == "Institution") or (i == "Location"):
        df2024[i] = df2024[i]
    else:    
        df2024[i] = pd.to_numeric(df2024[i])

df2024
```

```python
df2024.info()
```

```python
df2024.max()
```

```python
for i in df2024.columns:
    if i in ["Employability Rank", "Research Rank"]:
        # Apply where condition to specific columns, keeping values <= 1000 or setting to 1001
        df2024[i] = df2024[i].where(df2024[i] <= 1000, 1001)

df2024.max()
```

```python
# replace all nan values to max+1 value of their columns
for i in df2024.columns:
    if (i == "Institution") or (i == "Location"):
        df2024[i] = df2024[i]
    else:    
        df2024[i] = df2024[i].replace(np.nan, df2024[i].max() + 1)
df2024
```

```python
df2024.max()
```

```python
# Assuming df2014 is your DataFrame
cols = df2024.columns.tolist()  # Convert columns to a list

# Move 'Year' column to be right after 'Location' column
cols.insert(cols.index('Location') + 1, cols.pop(cols.index('Year')))

# Reorder DataFrame with the modified columns list
df2024 = df2024[cols]

# Display the updated DataFrame

df2024
```
