```python
import numpy as np
import pandas as pd
```

```python
df2017= pd.read_csv('2017.csv', sep='\t')
df2017
```

```python
# Remove rows where 'Institution' is NaN (assuming it indicates unnecessary rows)
df2017 = df2017.dropna(subset=['Institution'])
# Replace 'Top 0.*%' values in 'World Rank' column with sequential numbers
df2017['World Rank'] = df2017['World Rank'].replace(r'Top 0.*%', np.nan, regex=True)
df2017['World Rank'] = pd.Series(range(1, len(df2017) + 1), index=df2017.index)

# Reset index 
df2017 = df2017.reset_index(drop=True)

df2017
```

```python
df2017['Year']= 2017
df2017
```

```python
df2017.info()
```

```python
df2017.describe(include='all').T
```

```python
# Define the function to check for '>'
def has_more(value):
    return '>' in str(value)

df2017.applymap(has_more).sum()
```

```python
# Define the function to check for '+'
def has_plus(value):
    return '+' in str(value)

df2017.applymap(has_plus).sum()
```

```python
# Define the function to check for '-'
def has_dash(value):
    return '-' in str(value)

df2017.applymap(has_dash).sum()
```

```python
# Define the  '+' values with '0'
for i in df2017.columns:
    df2017[i] = df2017[i].replace(r'.*\+.*', '0', regex=True) 
    #r'.*\+.*' = regex pattern where .* matches any characters before and after +, effectively matching any string that contains +.
df2017
```

```python
# convert values of columns to int & float , except columns Institution & Location
for i in df2017.columns:
    if (i == "Institution") or (i == "Location"):
        df2017[i] = df2017[i]
    else:    
        df2017[i] = pd.to_numeric(df2017[i])

df2017
```

```python
df2017.describe().T
```

```python
df2017.max()
```

```python
# Replace all 0 values to max+1 value of their columns
for i in df2017.columns:
    if (i == "Institution") or (i == "Location"):
        df2017[i] = df2017[i]
    else:    
        df2017[i] = df2017[i].replace(0, df2017[i].max() + 1)
df2017
```

```python
df2017.describe().T
```

After 2019, the columns 'Publications,' 'Influence,' and 'Citations' were merged together to create a new column named 'Research Rank.
We created this new column by calculating the rounded mean of the three original columns.

```python
df2017['Research Rank'] = df2017[['Publications', 'Influence', 'Citations']].mean(axis=1).round().astype(int)
df2017.drop(columns=['Publications', 'Influence', 'Citations','Patents','Broad Impact'], inplace=True)
df2017
```

```python
# Assuming df2017 is your DataFrame
cols = df2017.columns.tolist()  # Convert columns to a list

# Move 'Year' column to be right after 'Location' column
cols.insert(cols.index('Location') + 1, cols.pop(cols.index('Year')))

# Move 'Research Rank' column to be right after 'Faculty Rank' column
cols.insert(cols.index('Faculty Rank') + 1, cols.pop(cols.index('Research Rank')))

# Reorder DataFrame with the modified columns list
df2017 = df2017[cols]

df2017
```
