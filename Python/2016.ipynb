```python
import numpy as np
import pandas as pd
```

```python
df2016= pd.read_csv('2016.csv', sep='\t')
df2016
```

```python
df2016['Year']= 2016
df2016
```

```python
df2016.describe(include='all').T
```

```python
# Define the function to check for '+'
def has_plus(value):
    return '+' in str(value)

df2016.applymap(has_plus).sum()
```

```python
# Define the function to check for '-'
def has_dash(value):
    return '-' in str(value)

df2016.applymap(has_dash).sum()
```

```python
# Define the function to check for '>'
def has_more(value):
    return '>' in str(value)

df2016.applymap(has_more).sum()
```

```python
# Define the  '+' values with '0'
for i in df2016.columns:
    df2016[i] = df2016[i].replace(r'.*\+.*', '0', regex=True) 
    #r'.*\+.*' = regex pattern where .* matches any characters before and after +, effectively matching any string that contains +.
df2016
```

```python
# convert values of columns to int & float , except columns Institution & Location
for i in df2016.columns:
    if (i == "Institution") or (i == "Location"):
        df2016[i] = df2016[i]
    else:    
        df2016[i] = pd.to_numeric(df2016[i])

df2016
```

```python
df2016.describe().T
```

```python
df2016.max()
```

```python
# replace all 0 values to max+1 value of their columns
for i in df2016.columns:
    if (i == "Institution") or (i == "Location"):
        df2016[i] = df2016[i]
    else:    
        df2016[i] = df2016[i].replace(0, df2016[i].max() + 1)
df2016
```

```python
df2016.describe().T
```

After 2019, the columns 'Publications,' 'Influence,' and 'Citations' were merged together to create a new column named 'Research Rank.
We created this new column by calculating the rounded mean of the three original columns.

```python
df2016['Research Rank'] = df2016[['Publications', 'Influence', 'Citations']].mean(axis=1).round().astype(int)
df2016.drop(columns=['Publications', 'Influence', 'Citations','Patents','Broad Impact'], inplace=True)
df2016
```

```python
# Assuming df2016 is your DataFrame
cols = df2016.columns.tolist()  # Convert columns to a list

# Move 'Year' column to be right after 'Location' column
cols.insert(cols.index('Location') + 1, cols.pop(cols.index('Year')))

# Move 'Research Rank' column to be right after 'Faculty Rank' column
cols.insert(cols.index('Faculty Rank') + 1, cols.pop(cols.index('Research Rank')))

# Reorder DataFrame with the modified columns list
df2016 = df2016[cols]

df2016
```
