```python
import numpy as np
import pandas as pd
```

```python
df2018= pd.read_csv('2018.csv', sep='\t')
df2018
```

```python
df2018['Year']= 2018
df2018
```

```python
# Define the function to check for '>'
def has_more(value):
    return '>' in str(value)

df2018.applymap(has_more).sum()
```

```python
# Define the function to check for '> 1000'
def has_more(value):
    return '> 1000' in str(value)

df2018.applymap(has_more).sum()
```

```python
# substitution of '> 1000' values with 1001 (maximum +1 ) using for loop
for i in df2018.columns:
    df2018[i] = df2018[i].replace(['> 1000'] , '1001') 

df2018.isin(['> 1000']).sum()
```

```python
# Define the function to check for '+'
def has_plus(value):
    return '+' in str(value)

df2018.applymap(has_plus).sum()
```

```python
# Define the function to check for '-'
def has_dash(value):
    return '-' in str(value)

df2018.applymap(has_dash).sum()
```

```python
# change - values to Nan
for i in df2018.columns:
    df2018[i] = df2018[i].replace(['-'] , np.nan)     

df2018
```

```python
# Define the function to check for '-'
def has_dash(value):
    return '-' in str(value)

df2018.applymap(has_dash).sum()
```

```python
# convert values of columns to int & float , except columns Institution & Location
for i in df2018.columns:
    if (i == "Institution") or (i == "Location"):
        df2018[i] = df2018[i]
    else:    
        df2018[i] = pd.to_numeric(df2018[i])

df2018
```

```python
df2018.describe().T
```

```python
df2018.max()
```

```python
# Replace all nan values to max+1 value of their columns
for i in df2018.columns:
    if (i == "Institution") or (i == "Location"):
        df2018[i] = df2018[i]
    else:    
        df2018[i] = df2018[i].replace(np.nan, df2018[i].max() + 1)
df2018
```

```python
df2018.info()
```

After 2019, the columns 'Publications,' 'Influence,' and 'Citations' were merged together to create a new column named 'Research Rank.
We created this new column by calculating the rounded mean of the three original columns.

```python
df2018['Research Rank'] = df2018[['Quality Publications', 'Influence', 'Citations']].mean(axis=1).round().astype(int)
df2018.drop(columns=['Quality Publications', 'Influence', 'Citations','Research Output'], inplace=True)
df2018
```

```python
# Assuming df2018 is your DataFrame
cols = df2018.columns.tolist()  # Convert columns to a list

# Move 'Year' column to be right after 'Location' column
cols.insert(cols.index('Location') + 1, cols.pop(cols.index('Year')))

# Move 'Research Rank' column to be right after 'Faculty Rank' column
cols.insert(cols.index('Faculty Rank') + 1, cols.pop(cols.index('Research Rank')))

# Reorder DataFrame with the modified columns list
df2018 = df2018[cols]
df2018
```
