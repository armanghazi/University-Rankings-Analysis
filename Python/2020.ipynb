```python
import numpy as np
import pandas as pd
```

```python
df2020= pd.read_csv('2020.csv', sep='\t')
df2020
```

```python
df2020['Year']= 2020
df2020
```

```python
df2020.info()
```

```python
df2020.describe(include='all').T
```

```python
# Define the function to check for '>'
def has_more(value):
    return '>' in str(value)

df2020.applymap(has_more).sum()
```

```python
# Define the function to check for '+'
def has_plus(value):
    return '+' in str(value)

df2020.applymap(has_plus).sum()
```

```python
# Define the function to check for '-'
def has_dash(value):
    return '-' in str(value)

df2020.applymap(has_dash).sum()
```

```python
# change - values to Nan
for i in df2020.columns:
    df2020[i] = df2020[i].replace(['-'] , np.nan)     

df2020
```

```python
# Define the function to check for '-'
def has_dash(value):
    return '-' in str(value)

df2020.applymap(has_dash).sum()
```

```python
# convert values of columns to int & float , except columns Institution & Location
for i in df2020.columns:
    if (i == "Institution") or (i == "Location"):
        df2020[i] = df2020[i]
    else:    
        df2020[i] = pd.to_numeric(df2020[i])
df2020
```

```python
df2020.info()
```

```python
df2020.max()
```

```python
# Replace all nan values to max+1 value of their columns
for i in df2020.columns:
    if (i == "Institution") or (i == "Location"):
        df2020[i] = df2020[i]
    else:    
        df2020[i] = df2020[i].replace(np.nan, df2020[i].max() + 1)
df2020
```

```python
df2020.max()
```

```python
for i in df2020.columns:
    if i in ["Employability Rank", "Research Performance"]:
        # Apply where condition to specific columns, keeping values <= 1000 or setting to 1001
        df2020[i] = df2020[i].where(df2020[i] <= 1000, 1001)

df2020.max()
```

```python
df2020.info()
```

```python
# Assuming df2020 is your DataFrame
cols = df2020.columns.tolist()  # Convert columns to a list

# Move 'Year' column to be right after 'Location' column
cols.insert(cols.index('Location') + 1, cols.pop(cols.index('Year')))
# Reorder DataFrame with the modified columns list
df2020 = df2020[cols]
df2020
```

```python
df2020.rename(columns={'Research Performance': 'Research Rank'}, inplace=True)
df2020
```
